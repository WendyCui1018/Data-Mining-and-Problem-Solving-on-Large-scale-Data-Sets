- (b) For each p and n = 1000, answer the following questions:
Are all random realizations of the ER network connected? Numerically
estimate the probability that a generated network is
connected. For one instance of the networks with that p, find
the giant connected component (GCC) if not connected. What
is the diameter of the GCC?

```r
is_connected(g1)
is_connected(g2)
is_connected(g3)
is_connected(g4)
is_connected(g5)
```

```r
largest_comp <- function(graph) {
cl <- components(graph)
V(graph)[which.max(cl$csize) == cl$membership]
}

GCC=largest_comp(g1)
Max_diameter=diameter(g1)
cat("For p=0.003, the Giant Connected Component is as follows.")
GCC
cat("For p=0.003, the diameter of the GCC = ",Max_diameter)
```

- (c) It turns out that the normalized GCC size (i.e., the size of the
GCC as a fraction of the total network size) is a highly nonlinear
function of p, with interesting properties occurring for
values where $p=O(\frac{ln(n)}{n})$. For $n = 1000$, sweep over values
of p in this region and create 100 random networks for each
p. Then scatter plot the normalized GCC sizes vs p. Empirically
estimate the value of p where a giant connected component
starts to emerge (define your criterion of emergence)?
Do they match with theoretical values mentioned or derived in
lectures?

```r
Max_p=log(N)/N
num=100
normalized_GCC <- rep(NA,num)
count=0
for (p in seq(length=num,from=1/N,to=Max_p)){
    count=count+1
    temp=rep(0,100)
    for (i in 1:100){
        g=random.graph.game(N,p)
        GCC=largest_comp(g)
        temp[i]=length(GCC)/N
    }
    normalized_GCC[count]=mean(temp)
}
xrange=seq(length=num,from=1/N,to=Max_p)
plot(xrange,normalized_GCC,xlab="probability p",ylab="normalized GCC size",main="Normalized GCC size v.s. p")
```

- (d) (i) Define the average degree of nodes $c = n \cdot p = 0.5$. Sweep
over number of nodes, n, ranging from 100 to 10000. Plot
the expected size of the GCC of ER networks with n nodes
and edge-formation probabilities $p = c/n$, as a function of n.
What trend is observed?

```r
c=0.5;
xrange=seq(100,1000,10)
expected_GCC_size=rep(0,length(xrange))
count=0;
for (n in xrange){
    count=count+1
    p=c/n;
    temp=rep(0,5000)
    for (i in 1:5000){
        g=random.graph.game(n,p)
        GCC=largest_comp(g);
        temp[i]=length(GCC)
    }
    expected_GCC_size[count]=mean(temp)
}
plot(xrange,expected_GCC_size,xlab="number of nodes: n",ylab="expected number of GCC size",main="E[GCC size] v.s. n with c=0.5")
```

- (ii) Repeat the same for $c=1$.

```r
c=1;
xrange=seq(100,1000,10)
expected_GCC_size=rep(0,length(xrange))
count=0;
for (n in xrange){
    count=count+1
    p=c/n;
    temp=rep(0,1000)
    for (i in 1:1000){
        g=random.graph.game(n,p)
        GCC=largest_comp(g);
        temp[i]=length(GCC)
    }
    expected_GCC_size[count]=mean(temp)
}
plot(xrange,expected_GCC_size,xlab="number of nodes: n",ylab="expected number of GCC size",main="E[GCC size] v.s. n with c=1")
```

- (iii) Repeat the same for values of $c = 1.1, 1.2, 1.3$, and show the
results for these three values in a single plot

```r
cc=c(1.1,1.2,1.3)
xrange=seq(100,1000,10)
expected_GCC_size=matrix(0,length(cc),length(xrange))
row=0
for (parameter in cc){
    row=row+1
    count=0;
    for (n in xrange){
        count=count+1
        p=parameter/n;
        temp=rep(0,500)
        for (i in 1:500){
            g=random.graph.game(n,p)
            GCC=largest_comp(g);
            temp[i]=length(GCC)
        }
        expected_GCC_size[row,count]=mean(temp)
    }
}

plot(xrange,expected_GCC_size[1,],col='green',xlab="number of nodes: n",ylab="expected number of GCC size",main="E[GCC size] v.s. n")
points(xrange,expected_GCC_size[2,],col='red')
points(xrange,expected_GCC_size[3,],col='blue')


```
